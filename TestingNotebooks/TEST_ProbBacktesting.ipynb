#### We create a return matrix (T, N)

T is the trading period.
N is the number of strategies.

```python
N = 100
T = 200
mu, sigma = 0, 0.01
M = np.random.normal(mu, sigma, size=(T, N))

print(M.shape)

S = 12

```

```python
print(sharpe_iid(M[:,1]))
```

```python
result = pbo(M, S, sharpe_iid, threshold=0.0, n_jobs=1, verbose=False)
print(result.pbo)
```
